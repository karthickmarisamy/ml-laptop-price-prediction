```python
import pandas as pd
```

```python
data = pd.read_csv('laptop_price - dataset.csv')
##data.sample(5)
```

```python
data.drop(columns=['Product'], inplace= True)
```

```python
## Split features X, y (independent and dependent features)


X = data.drop(columns=['Price (Euro)'])
y = data['Price (Euro)']
```

```python
X.head(1)
```

```python
X['new_screen'] = X['ScreenResolution'].str.split(' ')
```

```python
#X['new_screen1'] = X['new_screen'].map(lambda x : x if type(x) == int)
size = []
screen_name = []
for i in X['new_screen']:
   if len(i)-1 < 0:
      size.append("NA")
      screen_name.append("NA")
   else:
      size.append(i[len(i)-1])
      i.pop()
      if len(i) > 0:
         screen_name.append(' '.join(i))
      else:
         screen_name.append("NA")
      
X['screen_size'] = size
X['screen_name_new'] = screen_name
X
```

```python
X.sample(2)
```

```python
temp = X['screen_size'].str.split('x', n=1, expand=True)
X['screen_x_size'] = temp[0]
X['screen_y_size'] = temp[1]
```

```python
X.drop(columns=['screen_size', 'new_screen'], inplace=True)
```

```python
X.sample(2)
```

```python
X['Touch_screen'] = X['ScreenResolution'].apply(lambda x : 1 if 'Touchscreen' in x else 0)
```

```python
X['IPS'] = X['ScreenResolution'].apply(lambda x : 1 if 'IPS' in x else 0)
```
