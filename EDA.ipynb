```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('laptop_price - dataset.csv')
```

```python
data['Company'].value_counts()
```

```python
### Company based data count

plt.figure(figsize=(8,6))
sns.countplot(data=data, x='Company', palette=['#432371','#FAAE7B'], hue='Company')
plt.xlabel('Company name')
plt.ylabel('Company count')
plt.xticks(rotation=45)
plt.show()
```

```python
### Type Distribution

plt.figure(figsize=(6,7))
plt.pie(data['TypeName'].value_counts(), labels=data['TypeName'].value_counts().index, autopct='%1.1f%%')
plt.title('Type Distribution')
plt.show()
```

```python
## GPU company based distribution

plt.figure(figsize=(8,6))
plt.pie(data['GPU_Company'].value_counts(), labels=data['GPU_Company'].value_counts().index, autopct='%1.1f%%')
plt.title('GPU Company Distribution')
plt.show()
```

```python
## Company cpu distribution

plt.figure(figsize=(5,7))
plt.pie(data['CPU_Company'].value_counts(), labels=data['CPU_Company'].value_counts().index, autopct='%1.1f%%')
plt.title("Company CPU distribution")
plt.show()
```

```python
plt.figure(figsize=(5,8))
sns.countplot(data=data, x='OpSys', hue=data['OpSys'], palette=['#b02184', '#d39026', '#f6c03d', '#c14f37'])
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.title('Operating System count')
plt.show()
```

```python
plt.figure(figsize=(5,6))
sns.histplot(data['CPU_Frequency (GHz)'], color='#432371')
plt.xlabel('CPU_Frequency (GHz)')
plt.ylabel('Freqency')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data['ScreenResolution'], kde=True)
plt.xticks(rotation=90)
plt.xlabel('Screen Resolutions')
plt.ylabel('Resolution size')
plt.show()
```
